## Linear Regression: Checking Assumptions

Previously, we have demonstrated how to perform various useful tasks using linear regression models. These tasks include:

1. Calculating prediction intervals for our predicted *y* values
2. Calculating confidence intervals for our model parameters (slope, intercept)
3. Using *p*-values to assess our level of confidence in the existence of a relationship between *x* and *y*. 

However, all of these tasks rely upon certain assumptions about the data that we use to create our model. These assumptions are:

1. *Constant variance* - the amount of "noise" in our data should not change over the range of our dataset. 
2. *Normally-distributed Errors* - the residuals of our linear model should follow a *normal distribution*.
3. *Minimal distortion from outliers* - if outliers/leverage points are distorting our model, the model may not yield accurate predictions. 
4. *The underlying "true model" is linear* - there should be no evidence of a nonlinear relationship between *x* and *y*. 


If these assumptions do not hold, none of the results discussed above can be trusted. 
We will now teach you how to check that these assumptions are satisfied so that you can have confidence in your results.

## Using Graphs to Check Assumptions

Let's take a look at some of the data that we studied in earlier lessons:

```r
fuel.df <- read.csv("data/fuel-consumption.csv")
head(fuel.df)
```

A scatter plot of this data shows a strong linear trend:

```r
plot(
    FuelConsumedL ~ DistanceTraveledKM,
    data = fuel.df
)
```

We will build a linear model for this relationship:

```r
model <- lm(
    FuelConsumedL ~ DistanceTraveledKM,
    data = fuel.df
)
summary(model)
```

How do we know that we can trust the predictions from this model? R provides us with a convenient toolbox of graphs that allow us to check the validity of all relevant assumptions - we simply need to call `plot` on our `model` object to generate these four special graphs:

```r
opt.old <- par(no.readonly=TRUE)
par(mfrow=c(2,2))
plot(model)
par(opt.old)
```

Each of these four graphs can tell us if there is something wrong with our data that would make it inappropriate to use for construction of a linear model.
