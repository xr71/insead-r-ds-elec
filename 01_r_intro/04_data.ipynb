## Loading Data into R

In this lesson, you will get to experiment with loading some external data into R, which we will use to produce some rudimentary graphs. This is the first step on our journey towards working with more complex datasets. To load data from the file `cement.csv`, we will use the `read.csv` function - we'll store the data that we have loaded inside a variable named `cement.df`:

```r
cement.df <- read.csv("data/cement.csv")
```

Let's take a look at our data by viewing the contents of the variable `cement.df`: 

```r
cement.df
```

This data contains two columns - the GDP of the fictional country Fictitiousland (measured in trillion USD) and the cement imports of this same country (measured in billions USD) - we have used fictional data to spare you some of the complexities of real-world data for the time being.

We can see that this dataset is a more complicated type of object than we have worked with thus far. Specifically, the variable `cement.df` currently contains a *data frame*, which is a table-like structure similar to an excel spreadsheet. Like the vectors that we worked with in an earlier lesson, a dataframe contains many values inside a single object, but we can still store the entire object inside a single variable.

## Functions and Dataframes

What functions does R know that can be used to learn about our data? The first function to reach for when loading unfamiliar data is `summary`, which can be used to print a high-level analysis of our data frame:

```r
summary(cement.df)
```

We can see this function gives us useful information about each of the columns of our data frame; the minimum and maximum values of each column, the average value, etc. 

If we simply wish to count the number of rows and columns in our dataframe, we can use the `nrow` and `ncol` functions, respectively:

```r
print(paste("number of rows:", nrow(cement.df)))
print(paste("number of columns:", ncol(cement.df)))
```

Finally, we will want to make frequent use of the `head` function when first exploring a data frame. While we printed the entire data frame above simply by typing `cement.df`, in practice it might be prohibitively difficult to view the entire data frame at once for very large data sets with thousands or millions of rows. As such, usually we will examine unfamiliar data by passing the appropriate data frame to the `head` function, as shown below:

```r
head(cement.df)
```

By default, this will show the first `6` entries of the data frame - we can request a custom number using the named argument `n`:

```r
head(cement.df, n=3)
```

## Accessing Columns of Data

We will very frequently wish to pull out individual columns of data from a larger dataframe for various purposes. We can extract any column that we wish using the `$` operator, as shown below:

```r
cement.df$gdp.trillions
```

You can see above that using the `$` operator yields a *vector* of the column values, just like those we saw in previous lessons. For example, we can find the `sum` of all values in the `cement.imports.billions` column by extracting the appropriate column and passing the resulting vector to the `sum` function:

```r
sum(cement.df$cement.imports.billions)
```

## Plotting Data

When exploring new data, it is extremely important to graph the data appropriately to get a sense of the contents and patterns within a dataset. Let's  use a scatter plot to look at the relationship between `gdp.trillions` and `cement.imports.billions` - we'll use the named arguments `xlab` and `ylab` to label our axes appropriately, and use the `main` argument to give our graph a title:

```r
plot(
    cement.df$gdp.trillions,
    cement.df$cement.imports.billions,
    xlab="GDP ($trillion USD)",
    ylab="Cement ($billion USD)", 
    main="Cement Imports vs. GDP"
)
```

You can see that this function produces a straightforward scatter plot that shows a roughly linear relationship between GDP growth and increases in cement imports.
